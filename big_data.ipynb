```python
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv('credits.csv')
movies=pd.read_csv('movies_metadata.csv')
links=pd.read_csv('links.csv')
```

```python
movies=movies[['genres','id','title','overview','tagline']]
```

```python
movies['id'] = movies['id'].astype(str)
credits['id'] = credits['id'].astype(str)

movies = movies.merge(credits, on='id')
```

```python
movies
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
def convert2(text):
    if isinstance(text, list): 
        names = [i['name'] for i in text if isinstance(i, dict) and 'name' in i]
    else:
        try:
            data = ast.literal_eval(text)  
            names = [i['name'] for i in data if isinstance(i, dict) and 'name' in i]
        except (ValueError, SyntaxError, TypeError):
            return []

    return names[:3] 
```

```python
movies['cast'] = movies['cast'].apply(convert2)
movies.head()
```

```python
movies['crew'][1]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
keyword=pd.read_csv('keywords.csv')
```

```python
movies['id'] = movies['id'].astype(str)
keyword['id'] = keyword['id'].astype(str)

movies = movies.merge(keyword, on='id')
```

```python
movies
```

```python
movies.shape
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].fillna("").astype(str).apply(lambda x: x.split())
```

```python

movies['tagline'] = movies['tagline'].fillna("").astype(str).apply(lambda x: x.split())
```

```python

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']+movies['tagline']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew','tagline'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
    similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    
    if movie not in new['title'].values:
        print(f"❌ Movie '{movie}' not found in dataset.")
        return []

    index_list = new[new['title'] == movie].index
    if len(index_list) == 0:
        print(f"❌ Movie '{movie}' not found in dataset.")
        return []

    index = index_list[0]  
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    recommendations = []
    for i in distances[1:6]:  
        recommendations.append(new.iloc[i[0]]['title'])

    return recommendations
```

```python
recommend('Gandhi')
```
